Ques 1 Write a user-defined function that takes a number as input and returns:
• "Positive Even"
• "Positive Odd"
• "Negative"
• "Zero"
Use proper conditional statements and explain your logic

```python
def classify_number(n):
    if n == 0:
        return "Zero"
    elif n < 0:
        return "Negative"
    elif n % 2 == 0:
        return "Positive Even"
    else:
        return "Positive Odd"
```

Ques 3 Write a function that accepts a student’s marks and returns the grade based on the following logic:
• 90+ → A
• 75–89 → B
• 50–74 → C
• Below 50 → Fail
Use nested conditional statements and justify your approach.

```python
def get_grade(marks):
    if marks >= 50:
        if marks >= 75:
            if marks >= 90:
                return "A"
            else:
                return "B"
        else:
            return "C"
    else:
        return "Fail"
```

```python
print(get_grade(85))
```

Ques 4 Write a function that takes age as input.
• If age is negative, raise an error.
• If age is below 18, return "Minor", otherwise return "Adult".
Explain how conditional statements help in input validation

```python
def classify_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative.")
    elif age < 18:
        return "Minor"
    else:
        return "Adult"

```

```python
print(classify_age(15))
```

Ques 5 Write a function that accepts a list of integers and:
• Counts even numbers
• Ignores negative numbers
• Stops processing if zero appears
Use both loops and conditional statements inside the function

```python
def count_even_until_zero(numbers):
    even_count = 0

    for n in numbers:
        if n == 0:
            break    
        elif n < 0:
            continue      
        elif n % 2 == 0:
            even_count += 1

    return even_count
```

```python
print(count_even_until_zero([2, 4, -1, 3, 0, 6]))
```
